```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.externals.six import StringIO
```

```python
df = pd.read_fwf("../data/agaricus-lepiota.data", delimiter=',',
                 names=["classes", "cap-shape", "cap-surface",
                        "cap-color", "bruises", "odor",
                        "gill-attachment", "gill-spacing", "gill-size",
                        "gill-color", "stalk-shape", "stalk-root",
                        "stalk-surface-above-ring", "stalk-surface-below-ring", "stalk-color-above-ring",
                        "stalk-color-below-ring", "veil-type", "veil-color",
                        "ring-number", "ring-type", "spore-print-color",
                        "population", "habitat"])
df.head()
```

```python
df.describe()
```

```python
df = df.replace('?', np.nan)
```

```python
df.isnull().sum()
```

```python
df = df.drop(['stalk-root'], axis=1)
```

```python
df.shape
```

```python
letters_dict = {'a': 1,
  'b': 2,
  'c': 3,
  'd': 4,
  'e': 5,
  'f': 6,
  'g': 7,
  'h': 8,
  'i': 9,
  'j': 10,
  'k': 11,
  'l': 12,
  'm': 13,
  'n': 14,
  'o': 15,
  'p': 16,
  'q': 17,
  'r': 18,
  's': 19,
  't': 20,
  'u': 21,
  'v': 22,
  'w': 23,
  'x': 24,
  'y': 25,
  'z': 26,
   np.nan: np.nan}

def replace_letters(letter):
    if letter == np.nan:
        print(letter)
        return letter
    else:
        return letters_dict[letter]
```

```python
for item in df.columns:
    df[item] = df[item].apply(replace_letters)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['classes'].value_counts()
```

# Preprocessing

```python
X = df.drop(['classes'], axis=1)
y = df['classes']
y.head()
```

# Setting up Decision Tree

```python
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

# Modelling

```python
print('training set shape', X_trainset.shape, y_trainset.shape)
print('testing set shape', X_testset.shape, y_testset.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
mush_Tree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
mush_Tree 
```

```python
mush_Tree.fit(X_trainset, y_trainset)
```

# Prediction

```python
predTree = mush_Tree.predict(X_testset)
predTree[:5]
```

```python
print (predTree [0:5])
print (y_testset [0:5])
```

# Evaluation

```python
from sklearn import metrics
```

```python
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```
