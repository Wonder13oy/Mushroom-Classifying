# Mushroom Classifier

A mushroom, or toadstool, is the fleshy, spore-bearing fruiting body of a fungus, typically produced above ground, on soil, or on its food source. They tend to be divided into **Poisonous** or **Edible**. There are certain features that distinguish a mushroom into these catagories. For more information please follow this [link](https://en.wikipedia.org/wiki/Mushroom)

The aim is to use a decision tree model to predict whether mushrooms are poisonous or edible. This dataset includes descriptions of 23 species of gilled mushrooms in the Agaricus and Lepiota Family. Each species is identified as either edible or poisonous.

## Contents

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.externals.six import StringIO
```

## Data description

```python
df = pd.read_fwf("../data/agaricus-lepiota.data", delimiter=',',
                 names=["classes", "cap-shape", "cap-surface",
                        "cap-color", "bruises", "odor",
                        "gill-attachment", "gill-spacing", "gill-size",
                        "gill-color", "stalk-shape", "stalk-root",
                        "stalk-surface-above-ring", "stalk-surface-below-ring", "stalk-color-above-ring",
                        "stalk-color-below-ring", "veil-type", "veil-color",
                        "ring-number", "ring-type", "spore-print-color",
                        "population", "habitat"])
df.head()
```

```python
df.describe()
```

```python
df = df.replace('?', np.nan)
```

I found that there was missing data that was represented by `?`. I needed to replace this into something more conventional that is a `Nan`.

```python
df.isnull().sum()
```

```python
df = df.drop(['stalk-root'], axis=1)
```

```python
df.shape
```

I found that `stalk-root` had too many missing values for my liking, so I decided to drop the whole column. The missing values would not help in my predictions or analysis. We now move from 23 columns to 22, in doing so.

```python
df['classes'].value_counts()
```

This is the breakdown of the type of mushrooms found in the dataset. `e` is edible and `p` is poisonous.

# Preprocessing

```python
X = df.drop(['classes'], axis=1)
y = df['classes']
y.head()
```

```python
letters_dict = {'a': 1,
  'b': 2,
  'c': 3,
  'd': 4,
  'e': 5,
  'f': 6,
  'g': 7,
  'h': 8,
  'i': 9,
  'j': 10,
  'k': 11,
  'l': 12,
  'm': 13,
  'n': 14,
  'o': 15,
  'p': 16,
  'q': 17,
  'r': 18,
  's': 19,
  't': 20,
  'u': 21,
  'v': 22,
  'w': 23,
  'x': 24,
  'y': 25,
  'z': 26,
   np.nan: np.nan}

def replace_letters(letter):
    if letter == np.nan:
        print(letter)
        return letter
    else:
        return letters_dict[letter]
```

```python
for item in X.columns:
    X[item] = X[item].apply(replace_letters)

X.head()
```

The whole dataset contains alphabetic values. This does not sit well with Machine Learning algorithm. So I have decided to assign each letter of the alphabet to their corresponding positional number, i.e `a` being `1` and `z` being `26`. This still maintains the categorical nature of the data and this will change our predictions.

# Setting up Decision Tree

```python
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

I have split my data into a ratio of 7:3(train to test). This is to combat the problem of overfitting.

# Modelling

```python
print('training set shape', X_trainset.shape, y_trainset.shape)
print('testing set shape', X_testset.shape, y_testset.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
mush_Tree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
mush_Tree.fit(X_trainset, y_trainset)
```

# Prediction

```python
predTree = mush_Tree.predict(X_testset)
```

```python
print (f'Predicted results\n{predTree [0:5]}\n')
print (f'Actual results\n{y_testset [0:5]}')
```

I have ran test and displayed the first five predictions. When comparing these to the first five of the actual results, it got a 100% success rate. This is quite exciting so far, but we cannot judge the accuracy of a model from only the first five predictions. We need to compare to all actual results found.

# Evaluation

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
cm_df = pd.DataFrame(confusion_matrix(y_testset, predTree), columns=['Actual Edible','Actual Poisonus'],index=['Predicted Edible','Predicted Poisonous'])
cm_df
```

The above table shows the breakdown what model was able to predict correctly and what could be predicted incorrectly. The model was able to predict `1 275` mushrooms correctly to be edible, but predicted `4` incorrectly. It was able to correctly predict `1 073` mushrooms were poisonous but missed `86`. The ratios are quite acceptable and shows that the model did very well.

```python
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```

We have compared all our predictions to all the actual results. As you can see above, the model achieved above 90% accuracy score.
